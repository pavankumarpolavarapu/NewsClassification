## News Classification using Naive Bayes classifier

SKLEARN has a dataset called 20newsgroups which we are going to use as our dataset for classifying the news. Let's quickly load the required libraries and dataset.

```python
# Loading Required Libraries and Datasets

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

from sklearn.datasets import fetch_20newsgroups

data = fetch_20newsgroups()
print(type(data))
print(data.keys())
categories = data.target_names

train = fetch_20newsgroups(subset='train')
test  = fetch_20newsgroups(subset='test')
```

Now that we have the data, let us see what this data is all about.

```python
print(train.data[1])
```

Each data point is supposed to be an article but looks like an email. Hmm, that's fine.

Now let's import the classifiers and process the data.

```python
# We need Vectorizer to split the words from article and assign weight
from sklearn.feature_extraction.text import TfidfVectorizer

#Import Naive Bayes Multinomial classifier
from sklearn.naive_bayes import MultinomialNB

#Import pipeline for processing data through vectorizer and naive bayes classifier
from sklearn.pipeline import make_pipeline

model = make_pipeline(TfidfVectorizer(), MultinomialNB())

model.fit(train.data, train.target)

labels = model.predict(test.data)
```

Now that we have processed the information, let's see how the model behaved.

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(test.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', xticklabels=train.target_names, yticklabels=train.target_names)
plt.xlabel('Correct Label')
plt.ylabel('Predicted Label')
plt.rcParams['figure.figsize'] = [20, 10]
```

From the confusion matrix we can interpret that our model for most part got it right except for miscellaneous categories but overall it did a great job.

Let's extend this categorization beyond the articlees. Let's pass a string and try to find what our model thinks.

```python
def predict_category(s, train=train, model=model):
    label = model.predict([s])
    return train.target_names[label[0]]
```

```python
predict_category('Solar flares burst could potentially impact all electronic equipment')
```

```python
predict_category("The Senate prepared Wednesday to pass a short-term spending bill that would keep the government open through the New Year")
```
